```python
import torch 
import torch.nn as nn
```

```python
n, d_in, hidden, d_out = 32, 1000, 100, 10
x = torch.randn(n, d_in)
y = torch.randn(n, d_out)

class Model(nn.Module):
    def __init__(self, d_in, d_out, hidden):
        super(Model, self).__init__()
        self.linear1 = nn.Linear(d_in, hidden, bias = False)
        self.linear2 = nn.Linear(hidden, d_out, bias= False)
    def forward(self, x):
        y = self.linear1(x)
        y = self.linear2(y)
        return y
```

```python
model = Model(d_in, d_out, hidden)
loss_fn = nn.MSELoss(reduction= "sum")
lr = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr = lr)
```

```python
for e in range(1000):
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(e, loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
